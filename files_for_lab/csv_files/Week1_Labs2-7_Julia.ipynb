# Filling NaN

```python
#last column just have 1 value 'A', so NaN could be replaced with 'Not A'
dataset['vehicle type'] = dataset['vehicle type'].fillna('Not A')
```

```python
#state column mode is California so I'm going to try replacing NaN with the mode in a new df just in case
new_dataset = dataset.copy()
new_dataset['state'] = new_dataset['state'].fillna('California')
new_dataset['state'].unique()
```

```python
#response column I'm going to fill NaN with 'Not Sure' 
new_dataset['response'] = new_dataset['response'].fillna('Not Sure')
new_dataset['response'].unique()
```

```python
#vehicle class has 6 diferent options, so I'm going to fill NaN with 'Other'
new_dataset['vehicle class'] = new_dataset['vehicle class'].fillna('Other')
new_dataset['vehicle class'].unique()
```

```python
#vehicle size has 3 options: Smal, Large, Medsize (which is the mode) I'm going to replace NaN with the mode
new_dataset['vehicle size'] = new_dataset['vehicle size'].fillna('Medsize')
new_dataset['vehicle size'].unique()
```

```python
new_dataset.describe(include=[np.number]).T
```

```python
#number of open complaints has a median of 0 and it's also the mode, so I'm going to fill the NaN with 0
new_dataset['number of open complaints'] = new_dataset['number of open complaints'].fillna(0)
new_dataset.describe(include=[np.number]).T
```

It doesn't seem to afect very much to the mean and std so I think it's fine

```python
#months since last claim replace with the median 14
new_dataset['months since last claim'] = new_dataset['months since last claim'].fillna(14)
new_dataset.describe(include=[np.number]).T
```

I tried to replace it with the mean first, but I think the median distorts less all the data

```python
#response column I'm going to fill NaN with 'Not Sure' 
new_dataset['response'] = new_dataset['response'].fillna('Not Sure')
new_dataset['response'].unique()
```
