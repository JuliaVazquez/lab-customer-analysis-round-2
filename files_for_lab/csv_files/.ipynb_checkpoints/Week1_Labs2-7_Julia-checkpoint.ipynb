```python
#Replacing High class cars with Other

df2 = X_cat_new.replace({'Sports Car': 'Other', 'Luxury Car' : 'Other', 'Luxury SUV':'Other'})
df2
```

```python
from sklearn.preprocessing import OneHotEncoder

#defining a function to apply OneHotEncoder

def one_hot_encoder(col):
    col_df = pd.DataFrame(col)
    one_hot = OneHotEncoder(drop='first').fit(col_df)
    encoded_array = one_hot.transform(col_df).toarray()
#     print(encoded_array)
    col_names = col.unique()
    col_names = np.delete(col_names, [0])
#     print(col_names)
    df_encoded = pd.DataFrame(encoded_array, columns=col_names)
    return df_encoded
  
```

```python
#defining a loop to pass every column through the function and defining the new DataFrame

# new_df_names = []
new_df = normalized_data
for name in df2.columns:
#     a = 'encoded_' + name
#     print(a)
#     new_df_names.append(a)
    b = one_hot_encoder(df2[name])
    new_df = pd.concat([new_df, b], axis=1)
#     display(b)

new_df
```
